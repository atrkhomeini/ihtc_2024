**Room Skill Level**

Setiap pasien di rumah sakit memiliki informasi skill level minimum (skill_level_required) perawat yang mereka butuhkan di setiap shift pada masa tinggal mereka. Suatu kondisi dianggap sebagai penalty ketika skill level perawat tidak mencapai skill level minimum yang dibutuhkan pasien. Besaran penalti adalah selisih antara dua skill level tersebut.

*Notes: Jika skill level perawat yang ditugaskan di kamar pasien lebih tinggi dibandingkan skill level minimum yang dibutuhkan pasien, maka tidak dianggap sebagai penalty*

```python
data['patients'][0]
```

Logika utama:
1. Cek tiap penugasan perawat ke ruangan
2. Untuk tiap ruangan, cek semaua pasien dan occupants
3. Jika skill perawat kurang dari yang dibutuhkan, catat sebagai penalty
4. Akumulasi penalty dan kalikan dengan weight

```python
# Calculate room skill violations for each nurse assignment
total_violations = 0

# Process each nurse assignment
for nurse in sol['nurses']:
    # Get nurse skill level
    nurse_id = nurse['id']
    nurse_data = next(n for n in data['nurses'] if n['id'] == nurse_id)
    nurse_skill = nurse_data['skill_level']

    # Check each assignment
    for assignment in nurse['assignments']:
        day = assignment['day']
        shift_type = assignment['shift']
        shift = day * 3 + data['shift_types'].index(shift_type)

        # Check each room assigned to nurse
        for room_id in assignment['rooms']:
            # Check patients in room
            for patient in sol['patients']:
                if (patient['admission_day'] != 'none' and
                    patient['room'] == room_id):
                    # Get patient data
                    patient_data = next(p for p in data['patients'] if p['id'] == patient['id'])
                    admission_day = int(patient['admission_day'])
                    length_of_stay = patient_data['length_of_stay']

                    # Check if patient is present during this shift
                    start_shift = admission_day * 3
                    end_shift = start_shift + (length_of_stay * 3)

                    if start_shift <= shift < end_shift:
                        relative_shift = shift - start_shift
                        required_skill = patient_data['skill_level_required'][relative_shift]

                        if nurse_skill < required_skill:
                            violation = required_skill - nurse_skill
                            total_violations += violation
                            print(f"Nurse {nurse_id} (skill level {nurse_skill}) is underqualified for patient {patient['id']} in room {room_id} in shift {shift} (day{day}@{shift_type})")

            # Check occupants in room
            for occupant in data['occupants']:
                if occupant['room_id'] == room_id:
                    # Occupant starts from day 0
                    length_of_stay = occupant['length_of_stay']
                    end_shift = length_of_stay * 3

                    if shift < end_shift:
                        required_skill = occupant['skill_level_required'][shift]

                        if nurse_skill < required_skill:
                            violation = required_skill - nurse_skill
                            total_violations += violation
                            print(f"Nurse {nurse_id} (skill level {nurse_skill}) is underqualified for occupant {occupant['id']} in room {room_id} in shift {shift} (day{day}@{shift_type})")

print(f"\nTotal violations: {total_violations}")
print(f"Final penalty (with weight {data['weights']['room_nurse_skill']}): {total_violations * data['weights']['room_nurse_skill']}")
```

**Continuity of Care**

* Total perawat berbeda (distinct nurses) yang menangani pasien akan dihitung sebagai penalty dalam study case ini.
* Pada study case ini, rumah sakit mengharapkan setiap pasien dirawat oleh perawat yang sama selama masa tinggalnya. Namun, setiap perawat hanya berjaga 1 shift setiap harinya, sehingga setiap pasien akan dirawat oleh setidaknya 3 perawat berbeda selama masa tinggalnya.

*Notes: Occupant juga termasuk sebagai pasien di suatu rumah sakit*

```python
# Parse patient, occupant, and nurse assignments
patients = {p['id']: p for p in data['patients']}
nurses = {n['id']: n['assignments'] for n in sol['nurses']}
occupants = {o['id']: o for o in data['occupants']}
```

```python
# Calculate distinct nurses for each patient
continuity_of_care_patient = {}
for patient in sol['patients']:
    patient_id = patient['id']
    admission_day = patient.get('admission_day')
    room = patient.get('room')

    if admission_day == "none" or not room:
        # Skip unscheduled patients
        continue

    # Get patient's length of stay
    length_of_stay = patients[patient_id]['length_of_stay']

    # Find nurses assigned to this room during the stay
    distinct_nurses = set()
    for nurse_id, assignments in nurses.items():
        for assignment in assignments:
            if assignment['day'] in range(admission_day, admission_day + length_of_stay):
                if room in assignment['rooms']:
                    distinct_nurses.add(nurse_id)

    # Store the count of distinct nurses
    continuity_of_care_patient[patient_id] = len(distinct_nurses)
```

```python
# Calculate distinct nurses for each occupant
continuity_of_care_occupant = {}
for occupant in data['occupants']:
    occupant_id = occupant['id']
    admission_day = 0
    room = occupant.get('room_id')

    # Get patient's length of stay
    length_of_stay = occupants[occupant_id]['length_of_stay']

    # Find nurses assigned to this room during the stay
    distinct_nurses = set()
    for nurse_id, assignments in nurses.items():
        for assignment in assignments:
            if assignment['day'] in range(admission_day, admission_day + length_of_stay):
                if room in assignment['rooms']:
                    distinct_nurses.add(nurse_id)

    # Store the count of distinct nurses
    continuity_of_care_occupant[occupant_id] = len(distinct_nurses)
```

```python
# Print results
for occupant_id, nurse_count in continuity_of_care_occupant.items():
    print(f"Occupant {occupant_id} is cared for by {nurse_count} distinct nurses.")

for patient_id, nurse_count in continuity_of_care_patient.items():
    print(f"Patient {patient_id} is cared for by {nurse_count} distinct nurses.")
```

```python
# Calculate total
total_continuity = sum(continuity_of_care_patient.values()) + sum(continuity_of_care_occupant.values())
print(f"Total Continuity of Care Penalty: {total_continuity}")
print(f"Penalty Costs: {total_continuity * data['weights']['continuity_of_care']}")
```

Setiap pasien memiliki informasi beban kerja yang mereka hasilkan (workload_produced) di setiap shift selama masa tinggal mereka. Rumah sakit menginginkan perawat yang ditugaskan pada suatu shift tidak memiliki beban kerja yang melewati batas maksimum mereka. Suatu kondisi akan dianggap sebagai penalty ketika beban kerja perawat melewati batas maksimumnya.


```python
for patient in data['patients']:
    patient_id = patient['id']
    # Find the patient's data in sol['patients'] using patient_id
    patient_sol_data = next((p for p in sol['patients'] if p['id'] == patient_id), None)
    if patient_sol_data is not None:
            room = patient_sol_data.get('room')  # Use get to safely access 'room'
            admission_day = patient_sol_data['admission_day']
            length_of_stay = patient['length_of_stay']
    else:
            room = None  # Handle case where patient data is not found in sol
            admission_day = None
            length_of_stay = None
    workload_array = patient.get('workload_produced', [])  # Correct key: workload_produced
    print(f"{patient_id}, {room}, day{admission_day}, {length_of_stay}, {workload_array}")
```

```python
import copy
shifts = data['shift_types']
print(shifts)

# Create the 2D array with tripled days to represent shifts
#room_schedule = [[set() for _ in range(num_rooms)] for _ in range(days * len(shifts))]
room_schedule = [[{} for _ in range(num_rooms)] for _ in range(days * len(shifts))]
workload_schedule = copy.deepcopy(room_schedule)

print(workload_schedule)
```

```python
# Assume 'days', 'num_rooms', 'shifts', 'room_schedule', 'workload_schedule', 'sol', and 'data' are defined

for patient in sol['patients']:
    patient_id = patient['id']
    # Check if the 'room' key exists before accessing it
    room_id = patient.get('room')
    # If 'room' key doesn't exist, skip this patient and continue to the next
    if room_id is None:
        print(f"Warning: Patient {patient_id} has no assigned room. Skipping.")
        continue
    room_index = int(room_id[1:])  # Extract room index (e.g., 0 for "r0")
    admission_day = patient['admission_day']  # Get admission day

    # Find the corresponding patient data in data['patients']
    patient_data = next((p for p in data['patients'] if p['id'] == patient_id), None)

    if patient_data:
      length_of_stay = patient_data['length_of_stay'] # Pastikan length_of_stay diambil dari patient_data
      workload_produced = patient_data['workload_produced'] # Pastikan workload_produced diambil dari patient_data
      start_row_index = admission_day * len(shifts)  # Inisialisasi start_row_index di sini
      for day_offset in range(length_of_stay):
          for shift_index, shift in enumerate(shifts):
              current_row_index = start_row_index + day_offset * len(shifts) + shift_index

              # Periksa batas workload_schedule
              if 0 <= current_row_index < len(workload_schedule):
                  workload_value = workload_produced[day_offset * len(shifts) + shift_index]

                  # Access the room's workload dictionary for the current shift
                  room_workload_dict = workload_schedule[current_row_index][room_index]

                  # Update the workload count in the dictionary
                  room_workload_dict[workload_value] = room_workload_dict.get(workload_value, 0) + 1

                  # Debugging: Print nilai variabel
                  print(f"{patient_id}, day_offset: {day_offset}, shift {shift}, current_row: {current_row_index}, room_index: {room_index}, workload: {workload_value}, schedule: {workload_schedule[current_row_index][room_index]}")
```

```python
# ... (Assume 'days', 'num_rooms', 'shifts', 'room_schedule', 'workload_schedule', 'sol', and 'data' are defined) ...

# ... (Code for assigning workloads to workload_schedule) ...

# Add nurse IDs to workload_schedule
for nurse in sol['nurses']:
    nurse_id = nurse['id']
    for assignment in nurse['assignments']:
        day = assignment['day']
        shift = assignment['shift']
        rooms = assignment['rooms']

        # Calculate row index in workload_schedule
        row_index = day * len(shifts) + shifts.index(shift)

        # Add nurse ID to the corresponding rooms in the row
        for room_id in rooms:
            room_index = int(room_id[1:])  # Extract room index (e.g., 0 for "r0")

            # Check if workload_schedule already contain nurse
            # If not,  update the dictionary to store nurse ID as a key.
            if nurse_id not in workload_schedule[row_index][room_index]:
              workload_schedule[row_index][room_index][nurse_id] = []

# Print workload_schedule with nurse IDs
for day in range(days):
    print(f"Day {day}:")
    for shift_index, shift in enumerate(shifts):
        row_index = day * len(shifts) + shift_index
        row_data = workload_schedule[row_index]
        print(f"  {shift}: {row_data}")
```

```python
# ... (Assume 'days', 'num_rooms', 'shifts', 'room_schedule', 'workload_schedule', 'sol', and 'data' are defined) ...
# Check for nurse overwork
excessive_cost = 0
excessive_total = 0
excessive_weight = data["weights"]["nurse_eccessive_workload"]
for nurse in data['nurses']:
    nurse_id = nurse['id']
    for working_shift in nurse['working_shifts']:
        day = working_shift['day']
        shift = working_shift['shift']
        max_load = working_shift['max_load']
        total_workload = 0

        # Calculate row index in workload_schedule
        row_index = day * len(shifts) + shifts.index(shift)

        # Get the rooms assigned to the nurse for this shift
        assigned_rooms = []
        for i, room_set in enumerate(workload_schedule[row_index]):
            for key, value in room_set.items():
                if key == nurse_id:
                    assigned_rooms.append(i)
                    break  # Stop searching this room_set if nurse_id is found
        #print(assigned_rooms, nurse_id)

        # Calculate total workload for the nurse on this shift
        for room_index in assigned_rooms:
            room_data = workload_schedule[row_index][room_index]  # Get the room's data

            for workload, frequency in room_data.items():
                if isinstance(workload, int):  # Check if it's a workload value
                    total_workload += workload * frequency  # Multiply workload by frequency

        # Check for overwork
        if total_workload > max_load:
          excessive = total_workload - max_load
          print(f"Excessive workload {excessive} for nurse {nurse_id} in shift {row_index} (day{day}@{shift})")
          excessive_cost += excessive * excessive_weight
          excessive_total += excessive
```

**Surgeon Transfer**

Suatu kondisi dianggap sebagai penalty ketika dokter bedah ditugaskan pada ruang operasi yang berbeda dalam 1 hari

```python
# Menghitung jumlah dokter bedah
jumlahSurgeon = len(data['surgeons'])
jumlahSurgeon
```

```python
# Mengelompokkan pasien berdasarkan hari masuk

from collections import defaultdict
# Kelompokkan pasien berdasarkan hari masuk
patients_by_day = defaultdict(list)

for patient in sol["patients"]:
    admission_day = patient["admission_day"]
    if patient["admission_day"] != "none":
        patients_by_day[admission_day].append(patient)
for day, patients in sorted(patients_by_day.items()):
    print(f"Admission Day {day}:")
    for patient in patients:
        print(f"  - {patient}")
```

```python
# Menghitung jumlah operating_theater berbeda  per hari
distinct_operating_theaters_by_day = {}

for day, patients in patients_by_day.items():
    # Gunakan set comprehension untuk mendapatkan unique operating_theater
    distinct_operating_theaters = {patient["operating_theater"] for patient in
                                   patients}
    distinct_operating_theaters_by_day[day] = len(distinct_operating_theaters)

# Menampilkan hasil
print("Distinct Operating Theaters by Day:")
for day, count in sorted(distinct_operating_theaters_by_day.items()):
    print(f"  Day {day}: {count}")
```

```python
# Menghitung jumlah penalty surgeon transfer

surgeon_transfer_penalty = 0

for day, count in sorted(distinct_operating_theaters_by_day.items()):
    if count > jumlahSurgeon:
        surgeon_transfer_penalty += 1

print(f"SurgeonTransfer : {surgeon_transfer_penalty}")
```

**Patient Delay**

Patient Delay merupakan selisih waktu antara hari pasien diterima (admission date) dengan hari penerimaan paling awal yang mungkin bagi pasien (release date). Rumah sakit berharap nilai penalty ini dapat diminimalkan


Patient delay dapat dihitung sebagai:

*PatientDelay = admission_day - surgery_release_day*

```python
#Extract patients data from test data and solution data
patients = data["patients"]
sol_patients = sol["patients"]
```

```python
#Create dictionary for admission days
admission_days = {p["id"]: p["admission_day"] for p in sol_patients}
```

```python
admission_days
```

```python
#Initialize total patient delay
total_patient_delay = 0

#Iterate over patients
for patient in patients:
    patient_id = patient["id"]
    release_day = patient["surgery_release_day"]

    #Get the admission day from the solution data
    admission_day = admission_days.get(patient_id, "none")

    #Skip if admission_day is "none"
    if admission_day == "none":
        continue

    #Calculate delay
    delay = admission_day - release_day

    #Print individual patient delay
    if delay > 0:
        print(f"Patient {patient_id} has been delayed for {delay} days")

    total_patient_delay += delay

print(f"Total Patient Delay: {total_patient_delay} days")
```
