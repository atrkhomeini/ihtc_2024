#Penalty 1: RoomAgeMix

```python
# number of days
days = data ['days']

print(days)
```

```python
# number of rooms
num_rooms = len(data['rooms'])

print(num_rooms)
```

```python
# prompt: create two dimensional array of num_days and num_rooms consist of sets

room_age_mix = [[set() for _ in range(num_rooms)] for _ in range(days)]
room_age_mix
```

# Penalty 3: Continuity of Care

```python
# Parse patient, occupant, and nurse assignments
patients = {p['id']: p for p in data['patients']}
nurses = {n['id']: n['assignments'] for n in sol['nurses']}
occupants = {o['id']: o for o in data['occupants']}
```

```python
# Calculate distinct nurses for each patient
continuity_of_care_patient = {}
for patient in sol['patients']:
    patient_id = patient['id']
    admission_day = patient.get('admission_day')
    room = patient.get('room')

    if admission_day == "none" or not room:
        # Skip unscheduled patients
        continue

    # Get patient's length of stay
    length_of_stay = patients[patient_id]['length_of_stay']

    # Find nurses assigned to this room during the stay
    distinct_nurses = set()
    for nurse_id, assignments in nurses.items():
        for assignment in assignments:
            if assignment['day'] in range(admission_day, admission_day + length_of_stay):
                if room in assignment['rooms']:
                    distinct_nurses.add(nurse_id)

    # Store the count of distinct nurses
    continuity_of_care_patient[patient_id] = len(distinct_nurses)
```

```python
# Print results
for patient_id, nurse_count in continuity_of_care_patient.items():
    print(f"Patient {patient_id} is cared for by {nurse_count} distinct nurses.")
```

```python
# Calculate distinct nurses for each occupant
continuity_of_care_occupant = {}
for occupant in data['occupants']:
    occupant_id = occupant['id']
    admission_day = 0
    room = occupant.get('room_id')

    # Get patient's length of stay
    length_of_stay = occupants[occupant_id]['length_of_stay']

    # Find nurses assigned to this room during the stay
    distinct_nurses = set()
    for nurse_id, assignments in nurses.items():
        for assignment in assignments:
            if assignment['day'] in range(admission_day, admission_day + length_of_stay):
                if room in assignment['rooms']:
                    distinct_nurses.add(nurse_id)

    # Store the count of distinct nurses
    continuity_of_care_occupant[occupant_id] = len(distinct_nurses)
```

```python
# Print results
for occupant_id, nurse_count in continuity_of_care_occupant.items():
    print(f"Occupant {occupant_id} is cared for by {nurse_count} distinct nurses.")
```

```python
# Calculate total
total_continuity = sum(continuity_of_care_patient.values()) + sum(continuity_of_care_occupant.values())
print(f"Total Continuity of Care: {total_continuity}")
```

```python
# Calculate room skill violations for each nurse assignment
total_violations = 0

# Process each nurse assignment
for nurse in sol['nurses']:
    # Get nurse skill level
    nurse_id = nurse['id'] 
    nurse_data = next(n for n in data['nurses'] if n['id'] == nurse_id)
    nurse_skill = nurse_data['skill_level']
    
    # Check each assignment
    for assignment in nurse['assignments']:
        day = assignment['day']
        shift_type = assignment['shift']
        shift = day * 3 + data['shift_types'].index(shift_type)
        
        # Check each room assigned to nurse
        for room_id in assignment['rooms']:
            # Check patients in room
            for patient in sol['patients']:
                if (patient['admission_day'] != 'none' and 
                    patient['room'] == room_id):
                    # Get patient data
                    patient_data = next(p for p in data['patients'] if p['id'] == patient['id'])
                    admission_day = int(patient['admission_day'])
                    length_of_stay = patient_data['length_of_stay']
                    
                    # Check if patient is present during this shift
                    start_shift = admission_day * 3
                    end_shift = start_shift + (length_of_stay * 3)
                    
                    if start_shift <= shift < end_shift:
                        relative_shift = shift - start_shift
                        required_skill = patient_data['skill_level_required'][relative_shift]
                        
                        if nurse_skill < required_skill:
                            violation = required_skill - nurse_skill
                            total_violations += violation
                            print(f"Nurse {nurse_id} (skill level {nurse_skill}) is underqualified for patient {patient['id']} in room {room_id} in shift {shift} (day{day}@{shift_type})")
            
            # Check occupants in room 
            for occupant in data['occupants']:
                if occupant['room_id'] == room_id:
                    # Occupant starts from day 0 
                    length_of_stay = occupant['length_of_stay']
                    end_shift = length_of_stay * 3
                    
                    if shift < end_shift:
                        required_skill = occupant['skill_level_required'][shift]
                        
                        if nurse_skill < required_skill:
                            violation = required_skill - nurse_skill
                            total_violations += violation
                            print(f"Nurse {nurse_id} (skill level {nurse_skill}) is underqualified for occupant {occupant['id']} in room {room_id} in shift {shift} (day{day}@{shift_type})")

print(f"\nTotal violations: {total_violations}")
print(f"Final penalty (with weight {data['weights']['room_nurse_skill']}): {total_violations * data['weights']['room_nurse_skill']}")
```
